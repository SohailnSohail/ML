### Solution to task 

   #### a) train a linear regression model as a polynomial of degree 1

```python
# train a linear regression model
# Import necessary modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import random

# Fix the random seed
random.seed(10)

n_samples = 20  # Set the number of observations: training examples

x = np.sort(np.random.rand(n_samples))
true_y = np.cos(1.5 * np.pi * x)  # Define the underlying function
y = true_y + np.random.randn(n_samples) * 0.1  # Observations: noise has been added

X_test = np.linspace(0, 1, 100)  # The test data: values of the independent variable
Y_test = np.cos(1.5 * np.pi * X_test)  # TheÂ test data target from the underlying function

# Train linear regression model as a polynomial of degree 1
# Train linear regression model as a polynomial of degree 1
poly = PolynomialFeatures(degree=1)  # Create a polynomial features object with degree 1
X_poly = poly.fit_transform(x.reshape(-1, 1))  # Transform x data to include polynomial features
model = LinearRegression()  # Create a linear regression model
model.fit(X_poly, y)  # Train the model on the transformed data

# Make predictions on test data
y_test_pred = model.predict(poly.transform(X_test.reshape(-1, 1)))

```

```python
# RMSE
plt.plot(X_test, y_test_pred, label="Model")  # <- Corrected variable name
plt.plot(X_test, Y_test, label="True function")
plt.scatter(x, y, edgecolor="b", s=20, label="Samples")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim((0, 1))
plt.ylim((-2, 2))
plt.legend(loc="best")
plt.show()
```
