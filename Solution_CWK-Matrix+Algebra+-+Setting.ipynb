```python
""" Write the code to scatter plot the original data, as of the matrix above, as well as their transformation 
into a PCA based scatter plot with the covariance matrix being the transformation matrix """
import numpy as np
import matplotlib.pyplot as plt

# Original data
X = np.array([[-2, 2], [-4, -4], [4, 4], [2, -2], [0, 0]])

# Step 1: Compute the covariance matrix
cov_matrix = np.cov(X, rowvar=False)

# Step 2: Perform eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Step 3: Use the eigenvectors as the transformation matrix
transformed_data = X.dot(eigenvectors)

# Scatter plot
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA-based Scatter Plot')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Original data
X = np.array([[-2, 2], [-4, -4], [4, 4], [2, -2], [0, 0]])

# Step 1: Compute the covariance matrix
cov_matrix = np.cov(X, rowvar=False)

# Step 2: Perform eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Display eigenvalues and eigenvectors
print("Eigenvalues:")
for i in range(len(eigenvalues)):
    print("Eigenvalue {}: {:.4f}".format(i+1, eigenvalues[i]))

print("\nEigenvectors:")
for i in range(len(eigenvectors)):
    print("Eigenvector {}: {}".format(i+1, eigenvectors[:, i]))
```

#### Task 3
##### Modify and run the code for the PCA above, this time, however, by reducing the transformed space by one dimension. Subsequently, study write code to a) display the new data points, b) display the variance ratio being explained, c) display the PCA components as a matrix of eigenvectors, d) display the corresponding singular values.

### Solution of Task 3

```python
import numpy as np
import matplotlib.pyplot as plt

# Original data
X = np.array([[-2, 2], [-4, -4], [4, 4], [2, -2], [0, 0]])

# Step 1: Compute the covariance matrix
cov_matrix = np.cov(X, rowvar=False)

# Step 2: Perform eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Reduce dimensionality by one dimension
reduced_eigenvectors = eigenvectors[:, :-1]

# Step 3: Project the original data onto the reduced eigenvectors
transformed_data_reduced = X.dot(reduced_eigenvectors)

# Scatter plot of the reduced data to display new data points
plt.scatter(transformed_data_reduced[:, 0], np.zeros_like(transformed_data_reduced[:, 0]))
plt.xlabel('Principal Component 1')
plt.title('PCA-based Scatter Plot (Reduced Dimensionality)')
plt.grid(True)
plt.show()

# Display explained variance ratio 
explained_variance_ratio = eigenvalues[:-1] / np.sum(eigenvalues[:-1])
print("Explained Variance Ratio:", explained_variance_ratio)

# Display the PCA components (eigenvectors)
print("\nPCA Components (Eigenvectors):")
print(reduced_eigenvectors)

# Display the corresponding singular values
singular_values = np.sqrt(eigenvalues[:-1])
print("\nSingular Values:")
print(singular_values)
```

#### Task 4
##### Given the matrix below, decompose the matrix by applying SVD. In particular, you should calculate and display the matrices U, S and V by calculating eigenvalues and eigenvectors rather by simply applying the SVD function.

### Solution of Task 4

```python
import numpy as np

# Given matrix X
X = np.array([[-2, 0], [0, -1]])
print("X:")
print(X)

# Step 1: Compute X^T * X and X * X^T
XTX = np.dot(X.T, X)
XXT = np.dot(X, X.T)

# Step 2: Find the eigenvectors and eigenvalues of X^T * X and X * X^T
eigenvalues_XTX, U = np.linalg.eig(XTX)
eigenvalues_XXT, V = np.linalg.eig(XXT)

# Step 3: Compute matrices U, S, and V
# U and V are the eigenvectors
# S is the diagonal matrix containing the square roots of the eigenvalues
U = U[:, np.argsort(eigenvalues_XTX)[::-1]]  # Sort eigenvectors based on eigenvalues
V = V[:, np.argsort(eigenvalues_XXT)[::-1]]  # Sort eigenvectors based on eigenvalues
S = np.sqrt(np.diag(np.sort(eigenvalues_XTX)[::-1]))  # Sort eigenvalues and take square root to form S

# Display matrices U, S, and V
print("\nMatrix U:")
print(U)
print("\nMatrix S (Diagonal Singular Values):")
print(S)
print("\nMatrix V^T:")
print(V.T)
```
